```python
#Combien mange en moyenne un être humain ? Source => 

kcal_annuelles_humain = 2500 * 365
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
dispo_alimentaire = dispo_alimentaire.merge(population.loc[population['Année'] == 2017,["Zone", "Population"]],
                                            on='Zone')
```

```python
#Affichage du nouveau dataframe
dispo_alimentaire.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
dispo_alimentaire['dispo_kcal'] = dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_alimentaire['Population'] * 365
dispo_alimentaire['dispo_kcal'].sum()
```

```python
#Calcul du nombre d'humains pouvant être nourris
kcal_moy_annuel = 2500 * 365  

humains_nourris= dispo_alimentaire['dispo_kcal'].sum() / kcal_moy_annuel
print(humains_nourris)
```

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.8 - Evolution des 5 pays qui ont le plus bénéficiés de l'aide alimentaire entre 2013 et 2016</h3>
</div>

```python
#Création d'un dataframe avec la zone, l'année et l'aide alimentaire puis groupby sur zone et année 
data_filtered = aide_alimentaire[(aide_alimentaire['Année'] >= 2013) & (aide_alimentaire['Année'] <= 2016)]
#print(data_filtered.columns)
data_grouped = data_filtered.groupby(['Zone', 'Année'])['Aide_alimentaire'].sum().reset_index()
```

```python
#Création d'une liste contenant les 5 pays qui ont le plus bénéficiées de l'aide alimentaire
total_food_aid = data_grouped.groupby('Zone')['Aide_alimentaire'].sum().sort_values(ascending=False)


top_5_countries = total_food_aid.head(5).index.tolist()
top_5_countries
```

```python
#On filtre sur le dataframe avec notre liste
data_top_5 = data_grouped[data_grouped['Zone'].isin(top_5_countries)]
```

```python
# Affichage des pays avec l'aide alimentaire par année
for year in range(2013, 2017):
    display_year = data_top_5[data_top_5['Année'] == year]
    print(display_year)
```
